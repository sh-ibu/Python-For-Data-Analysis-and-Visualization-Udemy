# Split Apply and Combine

```python
# Create a function to apply
def ranker(df):
    df['acl_cont_rank']=np.arange(len(df))+1
    return df
```

```python
wine.sort('alcohol' , ascending=False,inplace=True)
```

```python
#wine
len(wine.groupby('quality'))
```

```python
wine=wine.groupby('quality').apply(ranker)
```

```python
wine[wine.acl_cont_rank==1]
```
