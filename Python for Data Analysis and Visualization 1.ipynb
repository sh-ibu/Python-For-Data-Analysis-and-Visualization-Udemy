# 13. Array Input and Output

```python
arr = np.array([1,2,3,4])
np.save('myarray.npy' , arr)
np.load('myarray.npy')
```

```python
arr1 = np.load('myarray.npy')
arr2 = np.array([2,4,5,6,7,8,9])
np.savez('ziparray' , x =arr1 , y=arr2)
```
